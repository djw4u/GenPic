###  Show the colorschemas and palettes

####  Imports and Constants

```python
# database to store pictures
DATABASE = 'sqlite:////Users/djw/data/gen_pic.sqlite'

# Display HTML and SVG output directly in the notebook 
from IPython.display import display, HTML

#Gen_Pic libaries 
from gp_base_db import *
from gp_sqlalchemy import *

#Initiliase SqlAlchemy
from sqlalchemy import create_engine
from sqlalchemy.orm import sessionmaker
engine = create_engine(DATABASE, echo=False)
Session = sessionmaker(bind=engine)
session = Session()
Base.metadata.create_all(engine)
```

```python
from string import Template
#Templates to generate the HTML page
html = """<html><head>
    <h5>Show colorschema: <b>$cs_name</b></h5>
    <style type="text/css">
        .circle { width:2%; padding:20px 20px;
        margin:0 auto; border:0px solid #a1a1a1;
        border-radius:20px; }
    </style>
    </head> <body> <table border="1">
    $rows
    </table> </body> </html>"""

row = "<tr> {0} </tr>"
d = """\t<div class="circle" style="background-color:{$n};"></div>"""
temp = Template(d)
```

Die Namen der Color Schemas: 
['corb31', 'corb50', 'all-corb50', 'smart02', 'svg-colors', 'all_corb31', 'BlackWhite']

```python
cs_name = 'svg-colors'
colors = session.query(Color_palette).filter_by(colorschema=cs_name)
len_of_palettes = len(colors[0].palette())

divs  = []
for i in range(1, len_of_palettes + 1):
    divs.append(temp.substitute(n=i))
col = """<td> {0} </td>\n<td>\n""" + '\n'.join(divs) + """\t</td>"""

rows, cols = [], []
for c in colors:
    if len(c.palette()) < len_of_palettes:
        continue
    cols.append(col.format(c.pname + '(' + str(c.id) + ')', *c.palette()))
    if len(cols) == 6:
        rows.append(row.format(' '.join(cols)))
        cols = []
if len(cols) > 0:
    rows.append(row.format(' '.join(cols)))

s = Template(html)
display(HTML(s.substitute(cs_name=cs_name , rows='\n'.join(rows))))
```
